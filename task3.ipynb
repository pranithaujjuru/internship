```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data3.csv')
data.head(10)
```

```python
plt.plot(data['own'],label='own')
plt.plot(data['year'],label='year')
plt.ylabel('own')
plt.xlabel('year')
plt.title('line chart')

plt.legend()
```

```python
#bar chart
plt.bar(data['year'],data['own'],color='cyan',label='records')
plt.xlabel('year')
plt.ylabel('own')
plt.title('bar graph')
plt.legend(loc='best')
```

```python
#scatter plot
plt.scatter(data['year'],data['own'])
plt.xlabel('year')
plt.ylabel('own')
plt.title('scatter plot')
plt.legend()
```

```python
#histograms
plt.hist(data['income'])
plt.xlabel('count')
plt.ylabel('income')
plt.title('histograms')
```
