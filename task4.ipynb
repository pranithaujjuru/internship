```python
import numpy as np
import pandas as pd 
import seaborn as sn
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df=pd.read_csv('USvideos.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
remove=['thumbnail_link','description']
df=df.drop(columns=remove)
df.info()
```

```python
import datetime 
```

```python
df['trending_date']=df['trending_date'].apply(lambda x :datetime.datetime.strptime(x,'%y.%d.%m'))
```

```python
df['trending_date']
```

```python
df['publish_time']=pd.to_datetime(df['publish_time'])
df.head(3)
```

```python
df['publish_month']=df['publish_time'].dt.month
df['publish_hour']=df['publish_time'].dt.hour
df['publish_day']=df['publish_time'].dt.day
df.head(3)
```

```python
print(sorted(df['category_id'].unique()))
```

```python
df['category_name']=np.nan
```

```python
df.loc[(df['category_id']==1),"category_name"]='Film and Animation'
df.loc[(df['category_id']==2),"category_name"]='Auto and Vehicles'
df.loc[(df['category_id']==10),"category_name"]='Music'
df.loc[(df['category_id']==15),"category_name"]='Pets and Animals'
df.loc[(df['category_id']==17),"category_name"]='Sports'
df.loc[(df['category_id']==19),"category_name"]='travel and events'
df.loc[(df['category_id']==20),"category_name"]='Gaming'
df.loc[(df['category_id']==22),"category_name"]='People and Blogs'
df.loc[(df['category_id']==23),"category_name"]='Comedy'
df.loc[(df['category_id']==24),"category_name"]='Entertainment'
df.loc[(df['category_id']==25),"category_name"]='News'
df.loc[(df['category_id']==26),"category_name"]='How to style'
df.loc[(df['category_id']==27),"category_name"]='Education'
df.loc[(df['category_id']==28),"category_name"]='Science and Technology'
df.loc[(df['category_id']==29),"category_name"]='non Profits and Activism'
df.loc[(df['category_id']==43),"category_name"]='Movies'
```

```python
df.head()
```

```python
df['year']=df['publish_time'].dt.year
df.head()
```

```python
yearly_count=df.groupby('year')['video_id'].count()
```

```python
yearly_count.plot(kind='bar',xlabel='year',ylabel='total publish count',title='total vedio publish per year')
plt.show()
```

```python
yearly_views=df.groupby('year')['views'].sum()
yearly_views.plot(kind='bar',xlabel='year',ylabel='toatal views',title='total views per year')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
category_views=df.groupby('category_name')['views'].sum().reset_index()
top=category_views.sort_values(by='views',ascending=False).head(5)
plt.bar(top['category_name'],top['views'])
plt.xlabel('Cateegory name',fontsize=12)
plt.ylabel('views',fontsize=12)
plt.title('top views',fontsize=12)
plt.xticks(rotation=100)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sn.countplot(x='category_name',data=df,order=df['category_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('vedio count by category')
plt.show()
```

```python
videos_perhour=df['publish_hour'].value_counts().sort_index()
plt.figure(figsize=(12,6))
sn.barplot(x=videos_perhour.index,y=videos_perhour.values,palette='crest')
plt.title('Number of videos published per hour')
plt.xlabel('hour')
plt.ylabel('number of videos')
plt.xticks(rotation=45)
plt.show()
```

```python
df['publish_date']=df['publish_time'].dt.date
```

```python
video_cunt_bydate=df.groupby('publish_date').size()
plt.figure(figsize=(12,6))
sn.lineplot(data=video_cunt_bydate)
plt.title('Number of videos published per day')
plt.xlabel('date')
plt.ylabel('number of videos')
plt.xticks(rotation=45)
plt.show()
```

```python
sn.scatterplot(data=df,x='views',y='likes')
plt.xlabel('views')
plt.ylabel('likes')
plt.title('views vs likes')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(2,2,1)
g=sn.countplot(x='comments_disabled',data=df)
g.set_title('Comments Disabled',fontsize=16)
g1=sn.countplot(x='ratings_disabled',data=df)
g1.set_title('ratings disabled',fontsize=16)
g2=sn.countplot(x='video_error_or_removed',data=df)
g2.set_title('video errors',fontsize=16)
plt.show()
```
