```python
df['target'].value_counts()
```

```python
df['target'].value_counts().plot(kind='bar',color=['cyan','red'])
lab=['yes','no']
plt.title("having disesase")
plt.xlabel('yes or no')
plt.ylabel("number")
plt.legend(["yes","No"])
```

```python
df['target'].value_counts().plot(kind='pie',figsize=(8,6))
plt.title("pie chart")
plt.legend(['Disease','NoDisease'])
```

```python
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title("classification according to gender")
plt.xlabel("0=heart disease and 1= heart disese")
plt.legend(0,1)
```

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie')
plt.title("Gender fraction")
plt.legend(['male','female'])
```

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar',color=['red','green','blue','black'])
plt.title("cancer type")
plt.legend(['tpe1','type2','type3','typ4'],loc='upper center')
```

```python
pd.crosstab(df.target,df.sex).plot(kind='bar',color=['green','blue','grey','cyan'])
plt.title("gender having cancer")
plt.legend(['Male',"female"])
```
